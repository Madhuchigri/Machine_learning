```python
import pandas as pd
```

```python
df = pd.read_csv('Test_Cricket_Players.csv')
df.head()
```

```python
df.drop(columns=[df.columns[0],df.columns[11]],inplace=True)
df.head()
```

```python
df.loc[ :, 'Runs' ] = df['Runs'].str.strip('*')
```

```python
df['Runs'] = pd.to_numeric(df['Runs'], errors='coerce')
df = df.dropna(subset=['Runs'])
df['Runs'].unique()
```

```python
df.head()
```

```python
type(df['Mins'])
```

```python
df = df.replace('-','0') # Replaced because there were values with '-' associated in the other columns
```

```python
df
```

```python
df[['Mins','BF','4s','6s','SR','Pos','Inns']] = df[['Mins','BF','4s','6s','SR','Pos','Inns']].apply(pd.to_numeric)
# converting the values to the numeric
```
